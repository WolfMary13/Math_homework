### Задание 3

Коробки с шоколадом упаковываются автоматически. Их средняя масса равна 1.06 кг. Известно, что 5% коробок имеют массу, меньшую 1 кг. Найдите: а) среднее квадратическое отклонение массы коробки, б) процент коробок, имеющих массу больше 1.1 кг.

```python
from scipy import stats
```

```python
# подобрадла вручную 
stats.norm.cdf(x=1, loc=1.06, scale=0.0365)
```

Логика подбора: подставила 1, потом 0,1 ... 0,01 ... 0,09 и т.д., сделала вывод, что между 0,04 и 0,03 подходящее значение и далее перебрала

Далее можно воспользоваться правилом для сигм, т.к. 1,06 + 0,0365 ~ 1,1 кг, следовательно больше 1,1 кг примерно 16% коробок
